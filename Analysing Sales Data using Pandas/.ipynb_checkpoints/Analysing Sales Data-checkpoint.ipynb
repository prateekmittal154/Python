# ANALYSING SALES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

### Task1 - Merge 12 csv files into 1

```python
df=pd.read_csv("./Sales_Data/Sales_April_2019.csv")

files=[file for file in os.listdir("./Sales_Data")]

all_months_data=pd.DataFrame()

for file in files:
    df=pd.read_csv("./Sales_Data/"+file)
    all_months_data= pd.concat([all_months_data,df])
    
all_months_data.to_csv("all_data.csv", index="False")
    
```

### Task 2- Read the updated CSV 

```python
all_data = pd.read_csv("all_data.csv")
all_data.head(10)
```

### Clean up the Data

```python
## First we Need to Clean up all The NOT A NUMERIC (NaN) values from our Data

all_data = all_data.dropna(how="any")
```

```python
# while typecasting date to int, there were errors popping up and this section of the code aims at clearing that


all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
```

```python
# We also need to convert each column to its correct data type!
# This is to avoid the error that we encountered while we tried (Quantity Ordered*Price Each)

all_data['Quantity Ordered'] = all_data['Quantity Ordered'].astype('int64')
all_data['Price Each']= all_data['Price Each'].astype('float64')
```

```python
del all_data['Unnamed: 0']
```

### Add Additional Fields to Our Data for Better Analysis 

```python
#Adding Month to Our Dataset

all_data['Month']=all_data['Order Date'].str[0:2]
all_data['Month']=all_data['Month'].astype('int32')
all_data.tail(10)
```

### Adding the col for total Sale Price for Each Product

```python
all_data['Sales']= all_data['Quantity Ordered']* all_data['Price Each']

all_data.head(20)
```

### Task-- > Add a column for city Name 

```python
all_data['City Name'] = all_data['Purchase Address'].apply(lambda x: x.split(',')[1])

all_data.head(20)
```

## Q1- which was the best month for sales? How much was earned that month?

```python
all_data.groupby(all_data['Month']).sum()['Sales']
```

```python
months= range(1,13)

Sales= [1.822257e+06,2.202022e+06,2.807100e+06,3.390670e+06,3.152607e+06,2.577802e+06,
        2.647776e+06,2.244468e+06,2.097560e+06,3.736727e+06,3.199603e+06,4.613443e+06]
plt.figure(figsize=(5,4), dpi=150)

plt.bar(months,Sales)
plt.title("Sales - Month Wise")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```

### Thus it can be concluded that the sales were maximum during the 12th month of the Year while January recorded the lowest Sales



##  Q2 Which City Had the Highest Amount of Sales?

```python
city_result= all_data.groupby(all_data['City Name']).sum()['Sales']
print(city_result)
```

```python
cities= [City for City , df in all_data.groupby('City Name')]

plt.figure(figsize=(5,4), dpi=150)
plt.bar(cities, city_result)
plt.xticks(rotation='vertical',size=8)
plt.title("Bar chart showing sales city wise")
plt.xlabel("CITY NAME")
plt.ylabel("Total Sales in USD")
plt.show()
```

## Q3- What time should I advertise the Product so that the maximum customers buy the product?

```python
all_data['Order Date']= pd.to_datetime(all_data['Order Date'])
```

```python
all_data['Hours']= all_data['Order Date'].dt.hour
all_data['Minute']= all_data['Order Date'].dt.minute
```

```python
hours=[hour for hour, df in all_data.groupby('Hours')]
plt.figure(figsize=(5,4), dpi=150)
plt.plot(hours,all_data.groupby(['Hours']).count())
plt.title("Number of Orders wrt Time")
plt.xlabel("Hours")
plt.ylabel("Number of Orders")
plt.xticks(hours, size=6)
plt.grid()
plt.show() 

## Using the plot that we have drawn using the above code, we can say that a good time to advertise maybe 11A.M to 12
## P.M and the again in the evening from 7:00 PM to 8:00 PM
```

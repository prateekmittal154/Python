```python
# importing the various necessary libraries 

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# read the data set into a data frame --> data

data=pd.read_csv("company_sales_data.csv")
```

### Q1- Read Total profit of all months and show it using a line plot

```python
monthNumber=data['month_number'].tolist()
TotalProfit=data['total_profit'].tolist()

plt.figure(figsize=(5,3),dpi=100)

plt.plot(monthNumber,TotalProfit, label='Month Wise Profit')
plt.title("Month wise Profit of the company")
plt.xlabel("Month Number")
plt.ylabel("Total Profit in a month")
plt.show()
```

### Q2 Read toothpaste sales data of each month and show it using a scatter plot


```python
ToothpasteData=data['toothpaste'].tolist()
plt.scatter(monthNumber,ToothpasteData)
plt.xlabel("Month Number")
plt.ylabel("Toothpaste sales for this month")
plt.grid(True, linewidth=0.2, linestyle="--")
plt.show()
```

# 2. Creating Threads


---



Threads are subsections of a single process which allow execution of the different sections of the code at the same time so that the total time taken for the program's execution is less and the CPU utilisation is maximum.



---







```python
import os, sys, time, threading, multiprocessing, random as r

startTime=time.time()

# Function to perform some task
def task(cmd):
    w=r.randint(10,20)
    print("....Task starts (running for %d... sec)"%(w))
    time.sleep(w)
    print("....Task ends")
    return

# Main Program
cmd=''
print("Program Started....")
print("..Thread starts")
t = threading.Thread(target=task , args=(cmd,))
t.start()
print("\n..Thread ends")

print("\nTotal Time %f sec"%(round(time.time() - startTime,4)))
print("Program Finished")
```

In the above code we observe that there is a small issue, after the execution of the main thread starts, it does not wait for the child threads to execute completely and hence, the program finishes early.

# 3. Basics of Multi-threading 

```python
startTime=time.time()

activeThreads = threading.activeCount()
print("Active Threads = ",activeThreads)

# Function to perform some task
def task(cmd):
    w=r.randint(10,20)
    print("....Task starts (running for %d sec)\n"%(w))
    time.sleep(w)
    print("....Task ends")
    return


# Main Program
print("Program Started....")
print("..Thread starts")

cmd=''
t = threading.Thread(target=task , args=(cmd,))
t.start()
	
# Waiting to finish all Threads
while True:
    if threading.activeCount() == activeThreads:
        break
    else:
        print ("...Thread Left %d..."%(threading.activeCount() - activeThreads))
        time.sleep(1)
  
print("All Threads compeleted")
print("\nTotal Time %f sec"%(round(time.time() - startTime,4)))
print("Program Finished")

```

# 4. Multi-threading - Waits for thread to compelete

```python
startTime=time.time()

activeThreads = threading.activeCount()
print("Active Threads = ",activeThreads)

# Function to perform some task
def task(cmd):
    w=r.randint(15,25)
    time.sleep(w)
    return


# Main Program - Run Multiple Threads
print("Program Started....")
for i in range(16):
    cmd=str(i+1)
    print("....Thread %s starts"%(cmd))
    t = threading.Thread(target=task , args=(cmd,))
    t.start()


# Waiting to finish all Threads
while True:
    if threading.activeCount() == activeThreads:
        break
    else:
        print ("...Thread Left %d..."%(threading.activeCount() - activeThreads))
        time.sleep(1)
  
print("All Threads compeleted")
print("\nTotal Time %f sec"%(round(time.time() - startTime,4)))
print("Program Finished")
```

# 5. Run Multiple Threads

```python
startTime=time.time()

activeThreads = threading.activeCount()
print("Active Threads = ",activeThreads)

numberOfCores=multiprocessing.cpu_count()
print("Number Of Cores = ",numberOfCores)

# Function to perform some task
def task(cmd):
    w=r.randint(5,15)
    time.sleep(w)
    return


# Main Program - Run Multiple Thread
print("Program Started....")
for i in range(16):
    cmd=str(i+1)
    print("....Thread %s starts"%(cmd))
    t = threading.Thread(target=task , args=(cmd,))
    t.start()
    while True:
        if threading.activeCount() - activeThreads + 1 <= numberOfCores:
            break
        time.sleep(1)


# Waiting to finish all Threads
while True:
    if threading.activeCount() == activeThreads:
        break
    else:
        print ("...Thread Left %d..."%(threading.activeCount() - activeThreads))
        time.sleep(1)
  
print("All Threads compeleted")
print("\nTotal Time %f sec"%(round(time.time() - startTime,4)))
print("Program Finished")
```

# 7. Create multiple Files one by one

```python
import random as r, string as s, shutil, os, time

startTime=time.time()

inputDirName='input'
totalFiles = 100
fileSize = 1*1024  # 887KB


try:
    # Delete input directory and then create it
    shutil.rmtree("./%s/"%(inputDirName))
    os.mkdir(inputDirName)
except:
    # Create the input directory
    os.mkdir(inputDirName)


# Function to perform some task    
def task(fileName):
    iteration=9000
    fileName="./%s/%s"%(inputDirName,fileName)
    while True:
        i=0
        fp=open(fileName,"a")
        for j in range(iteration):
            string=" ".join(r.sample(s.ascii_letters,50))
            fp.write(string + "\n")
        fp.close()
        if  os.path.getsize(fileName)/(fileSize) >= 50:
            break
    return None


# Main Program
print("Program Started....")
for i in range(totalFiles):
    fN="test%d.txt"%(i+1)
    task(fN)
    print ("Creating file %s"%(fN))
	
print("\nTotal Time %f sec"%(round(time.time() - startTime,4)))
print("Program Finished")
```

# 9. Create multiple Files using Multi-threading with Total threads = x * numberOfCores

```python

import random as r, shutil, string as s
import os, sys, time, threading, multiprocessing

startTime=time.time()

activeThreads = threading.activeCount()
print("Active Threads = ",activeThreads)

numberOfCores=multiprocessing.cpu_count()
print("Number Of Cores = ",numberOfCores)

# Num of Threads per core.
# Eg. if core=4, then total threads = 8
N = 2
print("Number Of Threds per Cores = %d\n"%(N))

inputDirName='input'
totalFiles = 100
fileSize = 1*1024  # 887KB

try:
	# Delete input directory and then create it
    shutil.rmtree("./%s/"%(inputDirName))
    os.mkdir(inputDirName)
except:
    # Create the input directory
    os.mkdir(inputDirName)

# Function to perform some task
def task(fileName):
    iteration=9000
    fileName="./"+inputDirName+"/"+fileName
    while True:
        i=0
        fp=open(fileName,"a")
        for j in range(iteration):
            string=" ".join(r.sample(s.ascii_letters,50))
            fp.write(string + "\n")
        fp.close()
        if  os.path.getsize(fileName)/(fileSize) >= 50:
            break
    return None


# Main Program - Run Multiple Thread
print("Program Started....")
for i in range(totalFiles):
    fN="test"+str(i+1)+".txt"
    t = threading.Thread(target=task , args=(fN,))
    t.start()
    print ("Thread started for ",fN)
    while True:
        if threading.activeCount() - activeThreads + 1 <= N * numberOfCores:
            break
        time.sleep(1)


# Waiting to finish all Threads
while True:
    if threading.activeCount() == activeThreads:
        break
    else:
        print ("...Thread Left %d..."%(threading.activeCount() - activeThreads))
        time.sleep(1)
  
print("All Threads compeleted")
print("\nTotal Time %f sec"%(round(time.time() - startTime,4)))
print("Program Finished")
```

# 10. Multiple File Processing one by one 

```python
import os, time, shutil

startTime=time.time()

inputDirName='input'
outputDirName='output'

# Check inputDirName directory exist or not
if os.path.exists(inputDirName) == False:
    print("!! Error: ", inputDirName, " directory doesn't exist")
    exit(0)


# Check outputDirName directory exist or not
try:
	# Delete output directory and then create it
    shutil.rmtree("./%s/"%(outputDirName))
    os.mkdir(outputDirName)
except:
    # Create the output directory
    os.mkdir(outputDirName)


# Function to perform some task - uppercase conversion
def task(fileName):
    inputputFileName="./"+inputDirName+"/"+fileName
    outputFileName="./"+outputDirName+"/"+fileName
    fpr=open(inputputFileName)
    fpw=open(outputFileName,"w")
	
    for line in fpr:
        fpw.write(line.upper())
    fpr.close()
    fpw.close()
    return None


# Main Program
print("Program Started....")
allFiles=os.listdir(inputDirName)
i=1
for fileName in allFiles:
    print ("File Processing %d (%s)"%(i,fileName))
    task(fileName)
    i=i+1
    
print("\nTotal Time %f sec"%(round(time.time() - startTime,4)))
print("Program Finished")
```
